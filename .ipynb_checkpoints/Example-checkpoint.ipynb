```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.real(1j**2)
```

```python
def z_wave(k_z,z):
    return np.real(np.exp(1j*k_z*z))
```

```python
k_z1 = -1.56e-2*(1+1j) # m^-1
k_z2 = 1.40e-4*(1-1j) # m^-1
k_z3 = 5.42e-6*(1) # m^-1
```

```python
z = np.arange(-300,0+0.5,0.5)
```

```python
plt.style.use("custom-light-mode")
```

```python
plt.figure(1,(4,6))
plt.plot(z_wave(k_z1,z),z,label='${k_z}^{(1)}$')
plt.plot(z_wave(k_z2,z),z,label='${k_z}^{(2)}$')
plt.plot(z_wave(k_z3,z),z,label='${k_z}^{(3)}$')
plt.xlabel('$\exp(ik_z z)$')
plt.ylabel('position $z$ (m)')
plt.ylim(z[0],z[-1])
plt.legend()
plt.title("Vertical wave\n components for perturbations")
plt.savefig("vert_wave_300.png",dpi=300)
```
